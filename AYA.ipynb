```python
excluded_columns = [
    "Clien_Name", "Secteur", "axe-reglement", "reglement", "reglement", "reglement", "CLE",
    "DATE_SAISIE", "DATE_ENREG", "DATE_MODIF", "TEMPS_SAISIE", "ORIGINE_SAISIE",
    "LANG_SAISIE", "APPAREIL_SAISIE", "PROGRESSION", "DERNIERE_QUESTION_SAISIE",
    "APPAREIL_SAISIE", "PROGRESSION", "DERNIERE_QUESTION_SAISIE"
]
```

```python
included_columns = [col for col in df_emi.columns if col not in excluded_columns]
```

```python
themes = set(col.split(' -')[0] for col in included_columns if '-' in col)
```

```python
themes
```

```python
for theme in themes:
    theme_cols = [col for col in included_columns if col.startswith(theme)]
    if theme_cols:
        df[theme + ' Moyenne'] = df[theme_cols].mean(axis=1)
```

```python
data_for_plot = df[[theme + ' Moyenne' for theme in themes]].mean().reset_index()
data_for_plot.columns = ['Theme', 'Moyenne']
```

```python
data_for_plot
```

```python
fig = px.bar(data_for_plot, x='Theme', y='Moyenne', title='Moyenne de Maturité par Thème', range_y=[0, 5])
```

```python
app = dash.Dash(__name__)

# Définition du layout de l'application
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

# Exécuter l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Création d'un diagramme de dispersion
fig_scatter = px.scatter(data_for_plot, x='Theme', y='Moyenne',
                         color='Moyenne',  # Utilise les moyennes comme couleur
                         size='Moyenne',  # Utilise aussi les moyennes pour la taille des points
                         title='Scatter Plot de Maturité par Thème')

# Ajouter ce graphique à Dash
scatter_layout = html.Div([
    dcc.Graph(figure=fig_scatter)
])
```

```python
# Création d'un box plot
fig_box = px.box(data_for_plot, y='Moyenne', x='Theme',
                 title='Box Plot de Maturité par Thème')

# Ajouter ce graphique à Dash
box_layout = html.Div([
    dcc.Graph(figure=fig_box)
])
```

```python
# Création d'un histogramme
fig_hist = px.histogram(data_for_plot, x='Moyenne',
                        nbins=20,  # Nombre de barres
                        title='Histogramme de Maturité')

# Ajouter ce graphique à Dash
hist_layout = html.Div([
    dcc.Graph(figure=fig_hist)
])
```

```python
# Création d'une matrice de corrélation pour l'exemple
correlation_matrix = df[[theme + ' Moyenne' for theme in themes]].corr()

# Création d'une heatmap
fig_heatmap = px.imshow(correlation_matrix,
                        labels=dict(x='Theme', y='Theme', color='Correlation'),
                        x=[theme + ' Moyenne' for theme in themes],
                        y=[theme + ' Moyenne' for theme in themes],
                        title='Heatmap de Corrélation entre Thèmes')

# Ajouter ce graphique à Dash
heatmap_layout = html.Div([
    dcc.Graph(figure=fig_heatmap)
])
```

```python
# Création d'un radar chart
fig_radar = px.line_polar(data_for_plot, r='Moyenne', theta='Theme',
                          line_close=True,
                          title='Radar Chart de Maturité par Thème')

# Ajouter ce graphique à Dash
radar_layout = html.Div([
    dcc.Graph(figure=fig_radar)
])
```

```python
app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Bar Chart', children=[html.Div(dcc.Graph(figure=fig))]),
        dcc.Tab(label='Scatter', children=[scatter_layout]),
        dcc.Tab(label='Box Plot', children=[box_layout]),
        dcc.Tab(label='Histogram', children=[hist_layout]),
        dcc.Tab(label='Heatmap', children=[heatmap_layout]),
        dcc.Tab(label='Radar Chart', children=[radar_layout])
    ])
])
```
