```python
import pandas as pd
import dash
from dash import html, dcc, Input, Output
import plotly.express as px
import dash_bootstrap_components as dbc
```

```python
df = pd.read_csv('data.csv', delimiter=';')

# Charger le fichier de mapping des noms de colonnes
mapping_df = pd.read_excel('Mapping_noms_variables_sphinx.xlsx')
```

```python
df.head()
```

```python
df
```

```python
column_mapping = dict(zip(mapping_df['Anciens Noms'], mapping_df['Nouveaux Noms']))
```

```python
mapping_df
```

```python
df.rename(columns=column_mapping, inplace=True)
```

```python
df.head()
```

```python
df_emi = df[df['Clien_Name'] == 'emi']
```

```python
df_emi
```

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
```

```python
app.layout = dbc.Container([
    dbc.NavbarSimple(
        children=[
            dbc.NavItem(dbc.NavLink("Page 1", href="/page-1")),
            dbc.NavItem(dbc.NavLink("Page 2", href="/page-2")),
            dbc.NavItem(dbc.NavLink("Page 3", href="/page-3"))
        ],
        brand="Dashboard Maturité",
        brand_href="/",
        color="primary",
        dark=True,
    ),
    dcc.Location(id='url', refresh=False),
    html.Div(id='page-content')
])
```

```python
@app.callback(Output('page-content', 'children'),
              [Input('url', 'pathname')])
def display_page(pathname):
    if pathname == '/page-1':
        return page_1_layout
    elif pathname == '/page-2':
        return page_2_layout
    elif pathname == '/page-3':
        return page_3_layout
    else:
        return "Bienvenue sur le Dashboard! Veuillez choisir une page."
```

```python
# Layouts pour chaque page
page_1_layout = dbc.Container([
    html.H1('Graphique 1'),
    dcc.Graph(
        figure=px.bar(df_emi, x='DATE_SAISIE', y='Gestion de continuité d\'activité - Aspects de la sécurité de l\'information en matière \nde gestion de la continuté d\'activité 1')
    )
])
```

```python
page_2_layout = dbc.Container([
    html.H1('Graphique 2'),
    dcc.Graph(
        figure=px.line(df_emi, x='DATE_SAISIE', y='Conformité - Conformité aux obligations légales et réglementaires 1')
    )
])
```

```python
page_3_layout = dbc.Container([
    html.H1('Graphique 3'),
    dcc.Graph(
        figure=px.scatter(df_emi, x='DATE_SAISIE', y='Conformité - Conformité aux obligations légales et réglementaires 1')
    )
])
```

```python
if __name__ == '__main__':
    app.run_server(debug=True)
```
